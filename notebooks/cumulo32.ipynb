```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Catálogos fusionados por Topcat
merged = pd.read_csv('match_completo', delimiter=',')
merged.head()
```

```python
cumulo = pd.DataFrame()
#FLujos SDSS, flujos LS, flujos wise, errores correspondientes:
#(En microJaskys y corregidos de extinción)
```

```python
#ra y dec de ambos catálogos
cumulo['ra_ls'] = merged['ra_1']
cumulo['ra_sdss'] = merged['ra_2']
cumulo['dec_ls'] = merged['dec_1']
cumulo['dec_sdss'] = merged['dec_2']
cumulo['type_ls'] = merged['type_1']
cumulo['type_sdss'] = merged['type_2']
```

```python
#Tengo las magnitudes sin corregir por extinvión (u,g,r,i,z) y corregidas (g,i,r,z)
#Añado dered_u: magnitud_X_corregida = magnitud_X - E(B-V)*coef(X)
coef_u = 3.995
coef_g = 3.214
coef_r = 2.165
coef_i = 1.592
coef_z = 1.211
e_bv = 0.113 #para mi cúmulo

#Magnitudes corregidas por extinción SDSS
cumulo['dered_mag_u_sdss'] = merged['modelMag_u']-coef_u*e_bv
cumulo['dered_mag_g_sdss'] = merged['modelMag_g']-coef_g*e_bv
cumulo['dered_mag_r_sdss'] = merged['modelMag_r']-coef_r*e_bv
cumulo['dered_mag_i_sdss'] = merged['modelMag_i']-coef_i*e_bv
cumulo['dered_mag_z_sdss'] = merged['modelMag_z']-coef_z*e_bv
```

```python
#Flujos corregidos por extinción SDSS
cumulo['dered_flux_u_sdss'] = 3631 * (10**6) * (10**(-0.4 * cumulo['dered_mag_u_sdss'])) #microJy
cumulo['dered_flux_g_sdss'] = 3631 * (10**6) * (10**(-0.4 * cumulo['dered_mag_g_sdss'])) #microJy
cumulo['dered_flux_r_sdss'] = 3631 * (10**6) * (10**(-0.4 * cumulo['dered_mag_r_sdss'])) #microJy
cumulo['dered_flux_i_sdss'] = 3631 * (10**6) * (10**(-0.4 * cumulo['dered_mag_i_sdss'])) #microJy
cumulo['dered_flux_z_sdss'] = 3631 * (10**6) * (10**(-0.4 * cumulo['dered_mag_z_sdss'])) #microJy
```

```python
#Errores de los flujos corregidos por extinción (microJy) SDSS
cumulo['err_dered_flux_u_sdss'] = cumulo['dered_flux_u_sdss'] * 0.4 * math.log(10) * merged['Err_u']
cumulo['err_dered_flux_g_sdss'] = cumulo['dered_flux_g_sdss'] * 0.4 * math.log(10) * merged['Err_g']
cumulo['err_dered_flux_r_sdss'] = cumulo['dered_flux_r_sdss'] * 0.4 * math.log(10) * merged['Err_r']
cumulo['err_dered_flux_i_sdss'] = cumulo['dered_flux_i_sdss'] * 0.4 * math.log(10) * merged['Err_i']
cumulo['err_dered_flux_z_sdss'] = cumulo['dered_flux_z_sdss'] * 0.4 * math.log(10) * merged['Err_z']
```

```python
#Magnitudes corregidas por extinción Legacy Survey
cumulo['dered_mag_g_ls'] = merged['dered_mag_g']
cumulo['dered_mag_i_ls'] = merged['dered_mag_i']
cumulo['dered_mag_r_ls'] = merged['dered_mag_r']
cumulo['dered_mag_z_ls'] = merged['dered_mag_z']
```

```python
#Flujos corregidos por extincón (nanomaggies) Legacy Survey
cumulo['dered_flux_g_ls'] = merged['dered_flux_g']
cumulo['dered_flux_i_ls'] = merged['dered_flux_i']
cumulo['dered_flux_r_ls'] = merged['dered_flux_r']
cumulo['dered_flux_z_ls'] = merged['dered_flux_z']
```

```python
#Flujos corregidos por extinción (microJy) Legacy Survey
cumulo['dered_flux_g_ls'] = cumulo['dered_flux_g_ls'] * 3.631
cumulo['dered_flux_i_ls'] = cumulo['dered_flux_i_ls'] * 3.631
cumulo['dered_flux_r_ls'] = cumulo['dered_flux_r_ls'] * 3.631
cumulo['dered_flux_z_ls'] = cumulo['dered_flux_z_ls'] * 3.631
```

```python
#Errores de los flujos corregidos por extinción (microJy) Legacy Survey
#error_f(microJy) = f(microJy) / (S/N) 
cumulo['err_dered_flux_g_ls'] = cumulo['dered_flux_g_ls']/merged['snr_g']
cumulo['err_dered_flux_i_ls'] = cumulo['dered_flux_i_ls']/merged['snr_i']
cumulo['err_dered_flux_r_ls'] = cumulo['dered_flux_r_ls']/merged['snr_r']
cumulo['err_dered_flux_z_ls'] = cumulo['dered_flux_z_ls']/merged['snr_z']
```

```python
#Flujos WISE (Legacy Survey)
#Paso de los flujos w1, w2, w3, w4 a microJy:
cumulo['flux_w1'] = merged['flux_w1'] * 3.631
cumulo['flux_w2'] = merged['flux_w2'] * 3.631
cumulo['flux_w3'] = merged['flux_w3'] * 3.631
cumulo['flux_w4'] = merged['flux_w4'] * 3.631
```

```python
#Error magnitudes WISE:
#error_mag = 2.5/(ln10*snr)
cumulo['err_mag_w1'] = 2.5 / (np.log(10) * merged['snr_w1'])
cumulo['err_mag_w2'] = 2.5 / (np.log(10) * merged['snr_w2'])
cumulo['err_mag_w3'] = 2.5 / (np.log(10) * merged['snr_w3'])
cumulo['err_mag_w4'] = 2.5 / (np.log(10) * merged['snr_w4'])
```

```python
#Error de los flujos WISE:
#error_f(microJy) = f(microJy) * 2.5* log10(e) * error_m 
cumulo['err_flux_w1'] = cumulo['flux_w1'] * 2.5 * np.log(10) * cumulo['err_mag_w1']
cumulo['err_flux_w2'] = cumulo['flux_w2'] * 2.5 * np.log(10) * cumulo['err_mag_w2']
cumulo['err_flux_w3'] = cumulo['flux_w3'] * 2.5 * np.log(10) * cumulo['err_mag_w3']
cumulo['err_flux_w4'] = cumulo['flux_w4'] * 2.5 * np.log(10) * cumulo['err_mag_w4']

cumulo.head()
```

```python
#Columna única de ra, dec (media de ra1, dec1 y ra2, dec2):
cumulo['ra'] = (merged['ra_1'] + merged['ra_2']) / 2
cumulo['dec'] = (merged['dec_1'] + merged['dec_2']) / 2
```

```python
ruta_archivo = 'cumulo32_completo.csv'
cumulo.to_csv(ruta_archivo, index=False)
print("DataFrame guardado en", ruta_archivo)
```
