```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Catálogos fusionados por Topcat
merged = pd.read_csv('match_completo', delimiter=',')
merged.head()
```

```python
#Tengo las magnitudes sin corregir por extinvión (u,g,r,i,z) y corregidas (g,i,r,z)
#Añado dered_u: magnitud_X_corregida = magnitud_X - E(B-V)*coef(X)
coef_u = 3.995
coef_g = 3.214
coef_r = 2.165
coef_i = 1.592
coef_z = 1.211
e_bv = 0.113 #para mi cúmulo

merged['dered_mag_u_sdss'] = merged['modelMag_u']-coef_u*e_bv
merged['dered_mag_g_sdss'] = merged['modelMag_g']-coef_g*e_bv
merged['dered_mag_r_sdss'] = merged['modelMag_r']-coef_r*e_bv
merged['dered_mag_i_sdss'] = merged['modelMag_i']-coef_i*e_bv
merged['dered_mag_z_sdss'] = merged['modelMag_z']-coef_z*e_bv

merged.head() 
```

```python
#Los flujos (g,i,r,z) dered_flux_X, flux_X están en nanomaggies. Bagpipes admite microJy por defecto
#Se multiplica por 3.631 para pasar a microJy
merged['flux_g_ls'] = merged['flux_g'] * 3.631
merged['flux_i_ls'] = merged['flux_i'] * 3.631
merged['flux_r_ls'] = merged['flux_r'] * 3.631
merged['flux_z_ls'] = merged['flux_z'] * 3.631

merged['dered_flux_g_ls'] = merged['dered_flux_g'] * 3.631
merged['dered_flux_i_ls'] = merged['dered_flux_i'] * 3.631
merged['dered_flux_r_ls'] = merged['dered_flux_r'] * 3.631
merged['dered_flux_z_ls'] = merged['dered_flux_z'] * 3.631

merged.head()
```

```python
#Me faltan flujos para la banda u. Los calculo a partir de las magnitudes 
merged['flux_u_sdss'] = 3631 * (10**6) * (10**(-0.4 * merged['modelMag_u'])) #microJy
merged['dered_flux_u_sdss'] = 3631 * (10**6) * (10**(-0.4 * merged['dered_mag_u_sdss'])) #microJy
merged.head()
```

```python
#Errores de los flujos (microJy)
merged['err_flux_u_sdss'] = merged['flux_u_sdss'] * 0.4 * math.log(10) * merged['Err_u']
merged['err_flux_g_ls'] = merged['flux_g_ls'] * 0.4 * math.log(10) * merged['Err_g']
merged['err_flux_r_ls'] = merged['flux_r_ls'] * 0.4 * math.log(10) * merged['Err_r']
merged['err_flux_i_ls'] = merged['flux_i_ls'] * 0.4 * math.log(10) * merged['Err_i']
merged['err_flux_z_ls'] = merged['flux_z_ls'] * 0.4 * math.log(10) * merged['Err_z']
merged.head()
```

```python
#Paso de los flujos w1, w2, w3, w4 a microJy:
merged['flux_w1'] = merged['flux_w1'] * 3.631
merged['flux_w2'] = merged['flux_w2'] * 3.631
merged['flux_w3'] = merged['flux_w3'] * 3.631
merged['flux_w4'] = merged['flux_w4'] * 3.631
```

```python
#necesito el error de las magnitudes:
#error_mag = 2.5/(ln10*snr)
merged['err_mag_w1'] = 2.5 / (np.log(10) * merged['snr_w1'])
merged['err_mag_w2'] = 2.5 / (np.log(10) * merged['snr_w2'])
merged['err_mag_w3'] = 2.5 / (np.log(10) * merged['snr_w3'])
merged['err_mag_w4'] = 2.5 / (np.log(10) * merged['snr_w4'])

#Errores de los flujos w1, w2, w3, w4:
#error_f(microJy) = f(microJy) * 2.5* log10(e) * error_m 
merged['err_flux_w1'] = merged['flux_w1'] * 2.5 * np.log(10) * merged['err_mag_w1']
merged['err_flux_w2'] = merged['flux_w2'] * 2.5 * np.log(10) * merged['err_mag_w2']
merged['err_flux_w3'] = merged['flux_w3'] * 2.5 * np.log(10) * merged['err_mag_w3']
merged['err_flux_w4'] = merged['flux_w4'] * 2.5 * np.log(10) * merged['err_mag_w4']

merged.head()
```

```python
#Columna única de ra, dec (media de ra1, dec1 y ra2, dec2):
merged['ra'] = (merged['ra_1'] + merged['ra_2']) / 2
merged['dec'] = (merged['dec_1'] + merged['dec_2']) / 2
```

```python
ruta_archivo = 'merged_manipulado.csv'
merged.to_csv(ruta_archivo, index=False)
print("DataFrame guardado en", ruta_archivo)
```
