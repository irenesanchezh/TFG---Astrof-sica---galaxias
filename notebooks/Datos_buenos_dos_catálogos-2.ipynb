```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
#Extraigo los datos de la web del Legacy Survey y compruebo que se han 
#guardado por columnas correctamente:
datos_web_LS = pd.read_csv('resultados_web_2_legacy_survey.txt', delimiter=',')
datos_web_LS.head()
```

```python
#Creo nuevas columnas para guardar los errores de las magnitudes:
datos_web_LS['err_mag_g'] = 2.5 / (datos_web_LS['snr_g'] * math.log(10))
datos_web_LS['err_mag_i'] = 2.5 / (datos_web_LS['snr_i'] * math.log(10))
datos_web_LS['err_mag_r'] = 2.5 / (datos_web_LS['snr_r'] * math.log(10))
datos_web_LS['err_mag_z'] = 2.5 / (datos_web_LS['snr_z'] * math.log(10))

datos_web_LS.head()
```

```python
#Los flujos dered_flux_X están en nanomaggies. Se multiplica por 3.631 para pasar a microJy)

datos_web_LS['dered_flux_g'] = datos_web_LS['dered_flux_g'] * 3.631
datos_web_LS['dered_flux_i'] = datos_web_LS['dered_flux_i'] * 3.631
datos_web_LS['dered_flux_r'] = datos_web_LS['dered_flux_r'] * 3.631
datos_web_LS['dered_flux_z'] = datos_web_LS['dered_flux_z'] * 3.631

datos_web_LS.head()
```

```python
#Con los flujos en microJy, calculo los error_flux_X = (flux_X)/(snr_X):

datos_web_LS['err_flux_g'] = (datos_web_LS['dered_flux_g'])/(datos_web_LS['snr_g'])
datos_web_LS['err_flux_i'] = (datos_web_LS['dered_flux_i'])/(datos_web_LS['snr_i'])
datos_web_LS['err_flux_r'] = (datos_web_LS['dered_flux_r'])/(datos_web_LS['snr_r'])
datos_web_LS['err_flux_z'] = (datos_web_LS['dered_flux_z'])/(datos_web_LS['snr_z'])

datos_web_LS.head()
```

```python
datos_web_SDSS = pd.read_csv('fotometria_cumulo_SDSS.txt', sep='\t')

datos_web_SDSS.head()
```

```python
#Las magnitudes modelMag_X no están corregidas de extinción:
#magnitud_X_corregida = magnitud_X - E(B-V)*coef(X)
coef_u = 3.995
coef_g = 3.214
coef_r = 2.165
coef_i = 1.592
coef_z = 1.211
e_bv = 0.113 #para mi cúmulo

datos_web_SDSS['dered_u'] = datos_web_SDSS['modelMag_u']-coef_u*e_bv
datos_web_SDSS['dered_g'] = datos_web_SDSS['modelMag_g']-coef_g*e_bv
datos_web_SDSS['dered_r'] = datos_web_SDSS['modelMag_r']-coef_r*e_bv
datos_web_SDSS['dered_i'] = datos_web_SDSS['modelMag_i']-coef_i*e_bv
datos_web_SDSS['dered_z'] = datos_web_SDSS['modelMag_z']-coef_z*e_bv

datos_web_SDSS.head() 
#Obtengo magnitudes más pequeñas (más brillantes). Los errores son los mismos que para model_Mag_X
```

```python
#Creo nuevas colummnas para los flujos (microJy) (calculados a partir de las magnitudes):
datos_web_SDSS['flux_u'] = 3631 * (10**6) * (10**(-0.4 * datos_web_SDSS['dered_u']))
datos_web_SDSS['flux_g'] = 3631 * (10**6) * (10**(-0.4 * datos_web_SDSS['dered_g']))
datos_web_SDSS['flux_r'] = 3631 * (10**6) * (10**(-0.4 * datos_web_SDSS['dered_r']))
datos_web_SDSS['flux_i'] = 3631 * (10**6) * (10**(-0.4 * datos_web_SDSS['dered_i']))
datos_web_SDSS['flux_z'] = 3631 * (10**6) * (10**(-0.4 * datos_web_SDSS['dered_z']))

datos_web_SDSS.head()   
```

```python
#Creo columnas nuevas para los errores de los flujos (microJy):
datos_web_SDSS['err_flux_u'] = datos_web_SDSS['flux_u'] * 0.4 * math.log(10) * datos_web_SDSS['Err_u']
datos_web_SDSS['err_flux_g'] = datos_web_SDSS['flux_g'] * 0.4 * math.log(10) * datos_web_SDSS['Err_g']
datos_web_SDSS['err_flux_r'] = datos_web_SDSS['flux_r'] * 0.4 * math.log(10) * datos_web_SDSS['Err_r']
datos_web_SDSS['err_flux_i'] = datos_web_SDSS['flux_i'] * 0.4 * math.log(10) * datos_web_SDSS['Err_i']
datos_web_SDSS['err_flux_z'] = datos_web_SDSS['flux_z'] * 0.4 * math.log(10) * datos_web_SDSS['Err_z']
datos_web_SDSS.head()
```

```python
#Compruebo que los flujos obtenidos del catálogo para LS-DR10 son los mismos que los obtenidos a partir de las magnitudes
#f(en microJy) = 3631E6 * 10^(-0.4*m) 

datos_web_LS['flux_g_calculado'] = 3631 * (10**6) * (10**(-0.4*datos_web_LS['dered_mag_g']))
datos_web_LS['flux_i_calculado'] = 3631 * (10**6) * (10**(-0.4*datos_web_LS['dered_mag_i']))
datos_web_LS['flux_r_calculado'] = 3631 * (10**6) * (10**(-0.4*datos_web_LS['dered_mag_r']))
datos_web_LS['flux_z_calculado'] = 3631 * (10**6) * (10**(-0.4*datos_web_LS['dered_mag_z']))

datos_web_LS.head()
#COINCIDEN LOS VALORES DERED_FLUX_X DEL CATÁLOGO CON CALCULADOS
```

```python
#Guardo los dataframes de los catálogos en archivos .csv

ruta_archivo = 'datos_LS.csv'
datos_web_LS.to_csv(ruta_archivo, index=False)  # No se especifica ningún separador para usar la coma por defecto
print("DataFrame guardado en", ruta_archivo)
```
